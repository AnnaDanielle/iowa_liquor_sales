# <span style="color:blue">Cleaning and Recategorizing the Data</span>

```python
import pandas as pd
import numpy as np
df_iowa = pd.read_csv('Iowa_Liquor_sales_sample_10pct.csv')
df_iowa.columns = ['date', 'store_num', 'city', 'zip', 'county_num', 'county', 'category', 'cat_name', 'vendor_num',
                  'item_num','item_descript', 'bottle_vol_ml', 'state_bottle_cost', 'state_bottle_retail',
                  'bottle_sold', 'sale', 'vol_sold_liter', 'vol_sold_gall']
```

```python
df_iowa.info()
```

```python
try:
    [int(i) for i in  df_iowa['zip']]
except:
    print i

df_iowa = df_iowa[df_iowa['zip'] != '712-2']
```

```python
df_iowa['zip'] = df_iowa['zip'].astype(int)
```

```python
df_iowa.head()
```

```python
# drop dollar signs, and make them floats
columns_dollar = ['state_bottle_cost','state_bottle_retail','sale']
for col in columns_dollar:
    df_iowa[col] = df_iowa[col].apply(lambda x: float(x[1:]))
```

```python
# reformat the date so it's not an object anymore, but now a datetime
df_iowa['date'] = pd.to_datetime(df_iowa['date'], format="%m/%d/%Y")
```

```python
# get rid of rows with nans
df_iowa = df_iowa.dropna()
```

```python
# calculate the dollar and percent of margin
df_iowa['margin_dollar'] = (df_iowa['state_bottle_retail'] - df_iowa['state_bottle_cost']) * df_iowa['bottle_sold']
df_iowa['margin_percent'] = (df_iowa['state_bottle_retail']/df_iowa['state_bottle_cost']) - 1
df_iowa['profit_per_gall'] = (df_iowa['margin_dollar']/df_iowa['vol_sold_gall'])
```

```python
# make the broad alcohol category
def make_broad_category(x):
    x = x.lower()
    if 'liqueur' in x or 'triple sec' in x:
        return 'liqueur'
    elif 'whisk' in x or 'scotch' in x or 'bour' in x or 'rye' in x:
        return 'whiskey'
    elif 'vodka' in x:
        return 'vodka'
    elif 'gin' in x:
        return 'gin'
    elif 'tequila' in x:
        return 'tequila'
    elif 'schnapps' in x:
        return 'schnapps'
    elif 'rum' in x:
        return 'rum'
    elif 'brandies' in x:
        return 'brandy'
    elif 'creme' in x:
        return 'creme'
    elif 'beer' in x:
        return 'beer'
    elif 'amaretto' in x:
        return 'amaretto'
    elif 'anisette' in x:
        return 'anisette'
    else:
        return 'misc'

df_iowa['broad_cat'] = df_iowa['cat_name'].apply(make_broad_category)
```

```python
check_misc = df_iowa[df_iowa['broad_cat'] == 'misc']
check_misc['cat_name'].unique()
```

```python
df_iowa.groupby('broad_cat', as_index = False).agg({'date': 'count'})
# probably should drop beer

df_iowa = df_iowa[df_iowa['broad_cat'] != 'beer']
```

```python
# create a pivot that will group each store number and the minimum and maximum date in the data
dates = df_iowa.groupby(by=['store_num'], as_index = False).agg({'date': ['min','max']})
dates.head()
```

```python
# create these arbitraty cutoffs even though you don't understand why you're choosing these particular dates
# but John told you to do this so you went along with it.
lower_cutoff = pd.Timestamp('20150115')
upper_cutoff = pd.Timestamp('20151231')
```

```python
# creating a mask that will only show the 'good stores', or the stores that were open during the majority of 2015
mask = (dates['date']['min'] < lower_cutoff) & (dates['date']['max'] > upper_cutoff)

# returns the boolean list of stores that fit the criteria
good_stores = dates[mask]['store_num']

# store this for later. this contains the entire dataset of stores that were open in 2015.
good_stores_df = df_iowa[df_iowa['store_num'].isin(good_stores)]
```

```python
# just pull 2015 data, maybe I'll only need 2015 at some point
good_stores_15_df = good_stores_df[good_stores_df['date'] < pd.Timestamp('20151231')]
```

```python
# summarize 2015 data
yearly_liquor_sales_by_store = good_stores_15_df.groupby('store_num', as_index = False).agg({'sale': 'sum',
                                                                                             'date': 'count',
                                                                                             'margin_percent': 'mean',
                                                                                             'margin_dollar': 'mean'
                                                                                         })
yearly_liquor_sales_by_store['avg_dollar_sales_per_day'] = (yearly_liquor_sales_by_store['sale']/
                                                            yearly_liquor_sales_by_store['date'])
yearly_liquor_sales_by_store.sort_values('avg_dollar_sales_per_day', ascending = False).head()
```

```python
# make the broad alcohol category
def alc_cat_margin_group(x):
    x = x.lower()
    if 'tequila' in x or 'whiskey' in x or 'brandy' in x or 'liqueur' in x:
        return 'high'
    elif 'rum' in x or 'vodka' in x or 'schnapps' in x or 'gin' in x:
        return 'medium'
    elif 'amaretto' in x or 'anisette' in x or 'creme' or 'misc' in x:
        return 'low'

good_stores_df_consol['margin_by_cat'] = good_stores_df_consol['broad_cat'].apply(alc_cat_margin_group)
```

# <span style="color:blue">Conclusion</span> 

It doesn't look like these features are helpful in determining where to open a store. If I were going to build this out for a client, I would request more data. I would be interested in traffic patterns, in the open/close time of stores in the vicinity, average family size, average age in the area. I might be interested in more seasonal data to predict where to open a store, like unemployment rate or yearly seasonality of sales, because it might tell a lot if a store outperforms the benchmark during a low unemployement period.

There are a couple of things that are clear that I would have expected without mining the data:

1) higher margins are in city centers where stores have more visibility to more people, and where markups are higher to subsize for higher rent.

2) tequila, whiskey and brandy have the highest profit margin.

3) there may be noise in terms of shelf real estate. For example, maybe some stores are more effective at pushing higher profit margin items.
